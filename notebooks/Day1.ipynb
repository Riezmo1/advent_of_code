# part 2

```python
numbers = ["eight","two","one","three","four","five", "six", "nine","seven"]
```

```python
for x in np.arange(0,len(data)):
    for ele in numbers:
        if ele in str(data[x]):
            data[x] = data[x].replace(ele,str(w2n.word_to_num(ele)))
```

```python
res = [ ''.join(filter(str.isdigit,str(x))) for x in data]
sum([int(x[0]+ x[-1]) for x in res])
```
